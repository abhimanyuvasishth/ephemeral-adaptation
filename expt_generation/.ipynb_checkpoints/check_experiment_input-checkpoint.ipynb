```python
import json
```

```python
expt_data = json.load( open( "experiment_data.json" , "rb" ) )
```

```python
print( "There are: " + str( len( expt_data ) ) )
```

```python
for key , value in expt_data[ 0 ].items():
    print( key )
    if key == "condition_order":
        print( "\t" + str( value ) )
    else:
        for key_2 , value_2 in value.items():
            print( "\t" + str( key_2 ) + " - " + str( value_2 ) )
```

```python
for expt in expt_data:
    for condition in expt[ "condition_order" ]:
        for block_type in [ "practice" , "experiment" ]:
            print( block_type + " - "  + condition )
            for block in expt[ block_type + "_blocks" ][ condition ]:
                print( block[ "words" ] )
                n_correct = 0
                for i in range( len( block[ "selection_locations" ] ) ):
                    if block[ "selection_locations" ][ i ] in block[ "predicted_locations" ][ i ]:
                        n_correct += 1
                print( "Accuracy: " + str( n_correct / float( len( block[ "selection_locations" ] ) ) ) + "\n" )
            
    print( "\n---\n" )
        
```
